```python
from sklearn.datasets import load_wine
import pandas as pd
```

```python
wine = load_wine()
```

```python
dir(wine)
```

```python
wine.feature_names
```

```python
wine.target
```

```python
wine.target_names
```

```python
df = pd.DataFrame(wine.data,columns = wine.feature_names)
df.head()
```

```python
df['target'] = wine.target
```

```python
df.head()
```

```python
df.groupby('target').describe()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(['target'],axis='columns'),wine.target,test_size=0.25)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model_G = GaussianNB()
model_G.fit(X_train,y_train)
```

```python
model_G.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
import numpy as np
```

```python
np.average(cross_val_score(GaussianNB(),wine.data,wine.target))
```

```python
np.average(cross_val_score(MultinomialNB(),wine.data,wine.target))
```
