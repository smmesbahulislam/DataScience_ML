```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/shahmdmesbahulislam/Documents/code/Data_Science_nd_ML/Machine Learning/carprices.csv")
df
```

```python
df['Car Model'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
feature_arr = ohe.fit_transform(df[['Car Model']]).toarray()
feature_arr
```

```python
ohe.categories_
```

```python
feature_labels = ohe.categories_
```

```python
import numpy as np
```

```python
feature_labels = np.array(feature_labels).ravel()
feature_labels
```

```python
after_ohe=pd.DataFrame(feature_arr,columns=feature_labels)
after_ohe
```

```python
merged = pd.concat([df,after_ohe],axis='columns')
merged
```

```python
final = merged.drop(['Car Model','Mercedez Benz C class'],axis='columns')
final
```

```python
X = final.drop('Sell Price($)',axis='columns')
X = X.values
X
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(X,y)
```
