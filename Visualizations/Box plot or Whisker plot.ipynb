# Box Plot / Whisker Plot :  
A standardized way to displaying the distribution of data based on a five number summary
  


Boxplot can tell you:  
* Tell you the values of your outliers
* Identify if data is symmetrical
* Determine how tightly data is grouped
* See if your data is skewed

### Load Data  
The data used to demonstrate boxplots is the Breast Cancer Wisconsin(Diagnostic) Data Set:  
http://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic) . The goal of the visualization is to show how the distributions for the column area_mean differs for benign versus malignant diagnosis.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/shahmdmesbahulislam/Documents/code/Data_Science_nd_ML/Datasets/wdbc.data",sep=',',header=None,names=['ID','Diagnosis','radius_mean','texture','perimeter','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave_points_mean','symmetry_mean','factal_diamension_mean','radius_SE','texture_SE','perimeter_SE','area_SE','smoothness_SE','compactness_SE','concavity_SE','concave_points_SE','symmetry_SE','factal_diamension_SE','radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave_points_worst','symmetry_worst','factal_diamension_worst'])
df.head()
```

```python
df.isnull().sum()
```

```python
# Looking at the distribution of the dataset in terms of diagnosis
df['Diagnosis'].value_counts(dropna = False)
```

### Plotting using Matplotlib

```python
malignant = df.loc[df['Diagnosis'] == 'M','area_mean'].values
benign = df.loc[df['Diagnosis']=='B','area_mean'].values
```

```python
plt.boxplot([malignant,benign],labels=['M','B'])
```

### Plotting using Pandas  
Pandas can be used as wrapper arount Matplotlib. One reason why you might want to plot using pandas is that it requires less code.  
We are going to create a boxplot to show how much less syntax you need to create the plot with pandas vs pure matplotlib.

```python
df.boxplot(column='area_mean',by='Diagnosis')
```

Sometimes you will find it useful to use Matplotlib syntax to adjust the final plot output. The code below removes the suptitle and title using pure matplotlib syntax.

```python
df.boxplot(column='area_mean',by='Diagnosis')
plt.title('')
plt.suptitle('')
```

### Plotting using Seaborn  
Seaborn can be seen as a wrapper on top of Matplotlib. Seaborn's website lists a bunch of advantages of using Seaborn including  
* Close integration with pandas data structures
* Dataset oriented API for examining relationships between multiple variables.
* Specialized support for using categorical variables to show observations or aggregate statistics.
* Concise control over matplotlib figure styling with several built-in themes.
* Tools for choosing color palettes that faithfully reveal patterns in your data.

```python
import seaborn as sns

sns.boxplot(x='Diagnosis',y='area_mean',data=df)
```
